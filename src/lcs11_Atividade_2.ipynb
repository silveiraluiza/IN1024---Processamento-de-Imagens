## Questão 1
### a)
Os componentes de um sistema de processamento de
imagens são: 



1.   Sensores de imagens (2 elementos);
2.   Hardware especializado;
3.   Um computador;
4.   Software de processamento de imagens;
5.   Capacidade de armazenamento em massa;
6.   Monitores de imagem;
7.   Sistemas de registro;
8.   Rede de comunicação. 



1. No ambito do sensoriamento, dois elementos são necessários para a aquisição de imagens digitais. Sendo o primeiro um dispositivo físico sensível à energia irradiada pelo objeto cuja imagem desejamos capturar. E o segundo um digitalizador, dispositivo utilizado para converter a saída do dispositivo físico de sensoriamento em formato digital. Por exemplo, em uma filmadora digital, os sensores produzem uma saída elétrica proporcional à intensidade da luz. O digitalizador converte essa saída em dados digitais. 

2. O hardware especializado em processamento de imagens normalmente consiste no digitalizador foi mencionado, além de um hardware capaz de desempenhar outras operações primárias, como uma unidade lógica e aritmética (ULA), que realiza operações aritméticas e lógicas em paralelo em toda a imagem. Uma ULA pode ser utilizada no cálculo da média de uma imagem à medida que ela é digitalizada, para realizar redução de ruídos. É uma unidade que realiza funções que requerem um rápido processamento de dados (como, por exemplo, a digitalização e o cálculo da média de imagens de vídeo a 30 quadros/s) que um computador comum não pode processar.

3. O computador em um sistema de processamento de imagens seria um computador de uso geral, que pode variar de um computador pessoal a um supercomputador. Praticamente qualquer computador pessoal bem equipado é suficiente para as tarefas de processamento de imagens off-line.

4. O software para o processamento de imagens consiste em módulos especializados que realizam tarefas específicas. Também inclui a possibilidade de o usuário escrever códigos que, no mínimo, utilizem os módulos especializados. Os pacotes de aplicativos mais sofisticados permitem a integração desses módulos e dos comandos gerais de software a partir de pelo menos uma linguagem computacional.

5. A capacidade de armazenamento em massa é indispensável em aplicações de processamento de imagens. Por exemplo, uma imagem do tamanho de 1.024 × 1.024 pixels, na qual a intensidade de cada pixel requer 8 bits, necessita de um espaço de armazenamento de 1 megabyte. Ao lidar com milhares, ou até milhões, de imagens, o armazenamento adequado em um sistema de processamento de imagens se torna um desafio. O armazenamento digital para aplicações de processamento de imagens se divide em três categorias principais: (1) armazenamento de curto prazo para utilização durante o processamento, (2) armazenamento online para acesso relativamente rápido e (3) armazenamento em arquivo para acesso com pouca frequência. 
  * Um método para proporcionar armazenamento de curto prazo é utilizar a memória do computador. Outro método é a utilização de frame buffers, que armazenam uma ou mais imagens e podem ser acessadas rapidamente, normalmente em velocidades de vídeo (por exemplo, em 30 imagens completas por segundo). Este método permite zoom de imagem praticamente instantâneo, além de scroll (deslocamentos verticais) e pan (deslocamentos ho- rizontais). O armazenamento online geralmente utiliza discos magnéticos ou mídias óticas. O principal fator que caracteriza o armazenamento online é o acesso frequente aos dados armazenados. Por fim, o armazenamento em arquivo é caracterizado por requisitos de armazenamento em massa, mas com baixa frequência de acesso. Fitas magnéticas e discos óticos alojados em “jukeboxes” são as mídias mais utilizadas para aplicações de arquivamento de imagens.

6. Os monitores de imagem utilizados hoje em dia são, em sua maioria, monitores de TV em cores (preferencial- mente de tela plana). Os monitores são controlados pelas placas de vídeo (gráficas ou de imagens), que são parte integral de um sistema computacional. Em alguns casos, é necessária a visualização estereoscópica (3-D), implementada a partir de um tipo de “capacete” contendo dois pequenos monitores de vídeo acoplados em um óculos, que deve ser utilizado pelo usuário.

7. Os sistemas de registro para as imagens incluem impressoras a laser, filmes fotográficos, impressora térmica, impressoras a jato de tinta e mídias digitais, como os discos óticos e de CD-ROM. O filme proporciona a mais alta resolução possível, mas o papel é o meio preferido para o material escrito. Para apresentações, as imagens são exibidas em transparências ou na forma digital se for utilizado um equipamento de projeção de imagens. A última abordagem está se tornando o padrão para apresentações de imagens.

8. A rede de comunicação é quase um componente padrão de qualquer sistema computacional em uso hoje em dia. Em razão do grande volume de dados inerente às aplicações de processamento de imagens, a princi- pal preocupação na transmissão de imagens é a largura de banda. Em redes dedicadas, isso normalmente não constitui um problema, mas as comunicações com sites remotos pela Internet nem sempre são eficientes.

Um exemplo de sistema de processamento de imagens é um sistema composto por uma impressora de digitalização (sensor de imagem, sistema de registro e hardware especializado) conectado a um computador pessoal (hardware espcializado, capacidade de armazenamento em massa, computador) onde uma folha contendo um desenho é digitalizada e visualizada em um monitor (monitores de imagem) então o desenho é tratado digitalmente por meio de um software de edição (Software de processamento de imagens), publicada na internet (rede de comunicação) e sua impressão em alta qualidade para enquadramento é realizada.


### bi) 
A subtração de imagens causa o realce de diferenças entre as imagens. Valores em preto na imagem de diferença indicam pontos nos quais não há nenhuma diferença entre as imagens das figuras. Ao subtrair repetidamente uma imagem de outra em algum ponto seria gerada a diferença absoluta entre as imagens.

### bii) 
A ordem de subtração importa.




```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import urllib.request as ur
from google.colab.patches import cv2_imshow
from matplotlib.pyplot import imshow

img_url1 = "https://thumbs.dreamstime.com/z/grupo-de-mancha-preta-da-aquarela-pinte-pontos-pretos-96897947.jpg"
img_url2 = "https://thumbs.dreamstime.com/z/manchas-e-pontos-do-preto-80702635.jpg"
# Baixando a imagem diretamente pela url e lendo com o opencv
def url_to_image(url):
  resp = ur.urlopen(url)
  image = np.asarray(bytearray(resp.read()), dtype="uint8")
  image = cv2.imdecode(image, cv2.COLOR_BGR2GRAY)
  return image


image1 = url_to_image(img_url1)
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
image2 = url_to_image(img_url2)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Diminuindo o tamanho da imagem

print('Original Dimensions 1: ',image1.shape)
print('Original Dimensions 2: ',image2.shape)  
scale_percent = 40 
width = int(image1.shape[1] * scale_percent / 100)
height = int(image1.shape[0] * scale_percent / 100)
dim = (width, height)

image1 = cv2.resize(image1, dim, interpolation = cv2.INTER_AREA)
image2 = cv2.resize(image2, dim, interpolation = cv2.INTER_AREA)

cv2_imshow(image1)
cv2_imshow(image2)
```

```
## Subtraindo as imagens 1 - 2
sub = image1.copy()
i = 0 
for i in range(10000):
  i = i +1
  sub = cv2.subtract(sub,image2) 

cv2_imshow(sub)

## Subtraindo as imagens 2 - 1
sub1 = image2.copy()
i = 0 
for i in range(10000):
  i = i +1
  sub1 = cv2.subtract(sub1,image1) 

cv2_imshow(sub1)
```

### c)

A interpolação é uma ferramenta básica utilizada extensivamente em tarefas como ampliação, redução, rotação e correções geométricas.
Ela é um processo que utiliza dados conhecidos para estimar valores em pontos desconhecidos. Suponha que uma imagem de tamanho 500 × 500 pixels deva ser ampliada uma vez e meia, resultando em uma imagem de 750 × 750 pixels (ampliação de 1,5 ×). Uma maneira simples de visuali- zar a ampliação é criar uma grade imaginária 750 × 750 com o mesmo espaçamento de pixels da imagem original e, então, reduzi-la de forma que se encaixe exatamente sobre a imagem original. Obviamente, o espaçamento de pixels na grade 750 × 750 após a redução será menor que o espaçamento de pixels na imagem original. Para realizar a atribuição de nível de intensidade para os pontos da imagem sobreposta, procuramos o pixel mais próximo na imagem original e atribuímos a intensidade desse pixel ao novo pixel na grade 750 × 750 reduzida. Quando terminamos de atribuir as intensidades a todos os pontos da grade sobreposta, nós a expandimos para o tamanho original a fim de obter a imagem ampliada. O mesmo método poderá ser aplicado caso se queira reduzir o tamanho da imagem, por exemplo de 500 x 500 para 300 x 300.

As interpolações podem utilizar mais que um vizinho mais próximo, sendo a bilinear, a que utiliza os quatro vizinhos mais próximos para estimar a intensidade de uma dada posição e a bicúbica a que inclui os 16 vizinhos mais próximos de um ponto.

Em geral, a interpolação bicúbica é melhor na preservação de detalhes finos em comparação com a interpolação bilinear. A interpolação bicúbica é o padrão utilizado em programas comerciais de edição de imagens, como o Adobe Photoshop e o Corel Photopaint. Apesar de terem perfomances melhores as interpolações mais complexas possuem um custo computacional maior.

### e)

A equalização de histograma é uma ferramenta adaptável de realce de contraste. Ela determina uma função de
transformação que busca produzir uma imagem de saída que tenha um histograma uniforme. Quando um realce automático é desejado, essa é uma boa abordagem porque os resultados dessa técnica são previsíveis, e sua implementação é fácil. 

### f)

```
def eq(image):
  hist,bins = np.histogram(image.flatten(),256,[0,256])
  cdf = hist.cumsum()
  cdf_m = np.ma.masked_equal(cdf,0)
  cdf_m = (cdf_m - cdf_m.min())*255/(cdf_m.max()-cdf_m.min())
  cdf = np.ma.filled(cdf_m,0).astype('uint8')
  return cdf[image]

i = 0
hist = image.copy()
while i < 100:
  hist = eq(hist)
  i += 1
  
print(image)
print(hist)
%matplotlib inline
imshow(np.asarray(hist))
```

```
img = image.copy()

def median_filter(data, filter_size):
    temp = []
    indexer = filter_size // 2
    window = [
        (i, j)
        for i in range(-indexer, filter_size-indexer)
        for j in range(-indexer, filter_size-indexer)
    ]
    index = len(window) // 2
    for i in range(len(data)):
        for j in range(len(data[0])):
            data[i][j] = sorted(
                0 if (
                    min(i+a, j+b) < 0
                    or len(data) <= i+a
                    or len(data[0]) <= j+b
                ) else data[i+a][j+b]
                for a, b in window
            )[index]
    return data


res = median_filter(img, 3)
laplacian = cv2.Laplacian(res,cv2.CV_64F)

cv2_imshow(image)
cv2_imshow(res)
cv2_imshow(laplacian)
```

Como o influxo de brilho durante o dia é normal, a normalização só deve ser aplicada no horário da noite, obtendo-se um valor de referência para a iluminação artificial quando não houver objetos em movimento no frame da câmera para que não interfiram na iluminação, esse frame estático noturno será o valor pela qual todas as outras imagens serão normalizadas.

Podemos lidar com isso selecionando várias subáreas representativas do plano da imagem que tem menor probabilidade de sofrerem alteração na iluminação por conta de movimento e computar sua intensidade média, em seguida, selecionar o mínimo e o máximo de todos os valores médios individuais, indicados por, fmin e fmax. 

O objetivo, então, é processar qualquer imagem de entrada, f (x, y), de modo que seu mínimo e seu valor máximo sejam iguais a min e fmax, respectivamente. Uma maneira de fazer isso é com uma função de transformação linear `f(x, y) = a f(x, y) + b` onde a saída é a imagem modificada.
